```python
import requests
import random
import time

url = 'https://raw.githubusercontent.com/CriMenghini/ADM/master/2019/Homework_3/data/movies3.html'
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
lst = []
for link in soup.find_all('a'):
    lst.append(link['href'])
```

```python
path = 'C:/Users/melik/Desktop/DS'
if type(path)!=str:
    raise Exception('input path must be a string! Not a {}'.format(type(path)))
    
i = 20000
for l in lst:
    with open (''.join([path, 'article_', str(i), 'html']), 'w') as f:
        response1 = requests.get(l)
        soup1 = BeautifulSoup(response1.text, 'html.parser')
        soup1.body.find_all('a')
        try:
            txt = requests.get(l.get('href')).text
        except:
            time.sleep(2*60)
            txt = requests.get(l.get('href')).text
        f.write(txt)
    i += 1
```
